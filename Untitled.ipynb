```python
def lot_fix (c1, c2, c3, c4, c5):
    if 'Lot' in str(c1):
        return c1[9:-4]
    if 'Lot' in str(c2):
        return c2[9:-4]
    if 'Lot' in str(c3):
        return c3[9:-4]
    if 'Lot' in str(c4):
        return c4[9:-4]
    if 'Lot' in str(c5):
        return c5[9:-4]
    else:
        return "-1"
```

```python
def built_fix(c1, c2, c3, c4, c5):
    if 'Built' in str(c1):
        return c1
    if 'Built' in str(c2):
        return c2
    if 'Built' in str(c3):
        return c3
    if 'Built' in str(c4):
        return c4
    if 'Built' in str(c5):
        return c5
    else:
        return "-1"
```

```python
df['Lot Size'] = df.apply(lambda x: lot_fix(x.SSeMC, x.Neighborhood, x.cERLyX, x.Field2, x.Field3), axis = 1)
```

```python
df['Age of Building'] = df.apply(lambda x: built_fix(x.SSeMC, x.Neighborhood, x.cERLyX, x.Field2, x.Field3), axis = 1)
```

```python
df['Age of Building'] = df['Age of Building'].apply(lambda x: x.split(' ')[-1])
```
